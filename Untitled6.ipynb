```python
import numpy as np

class ANFIS:
    def __init__(self, n_inputs=None, n_rules=10, learning_rate=0.01):
        self.n_rules = n_rules
        self.learning_rate = learning_rate
        self.n_inputs = n_inputs

        # Initialize membership function parameters randomly
        if n_inputs is not None:
            self.mf_params = np.random.rand(n_inputs, n_rules, 2)
        else:
            self.mf_params = None

        # Initialize rule weights and consequent parameters randomly
        self.rule_weights = np.random.rand(n_rules)
        self.consequent_params = np.random.rand(n_rules)

    def forward_pass(self, X):
        # Forward pass through the ANFIS
        batch_size = X.shape[0]

        # Calculate membership degrees
        memberships = np.zeros((batch_size, self.n_rules, self.n_inputs))
        for i in range(self.n_rules):
            for j in range(self.n_inputs):
                memberships[:, i, j] = self.gaussian_mf(X[:, j], self.mf_params[j, i])

        # Normalize membership degrees
        normalized_memberships = memberships / np.sum(memberships, axis=1, keepdims=True)

        # Calculate firing strengths
        firing_strengths = np.prod(normalized_memberships, axis=2)

        # Normalize firing strengths
        normalized_firing_strengths = firing_strengths / np.sum(firing_strengths, axis=1, keepdims=True)

        # Reshape consequent parameters
        consequent_params_reshaped = self.consequent_params.reshape(1, -1)

        # Transpose consequent_params_reshaped to have shape (n_rules, 1)
        consequent_params_transposed = consequent_params_reshaped.T

        # Calculate weighted consequents
        weighted_consequents = normalized_firing_strengths.dot(consequent_params_transposed)

        print("Shape of self.consequent_params:", self.consequent_params.shape)
        print("Shape of consequent_params_transposed:", consequent_params_transposed.shape)
        print("Shape of weighted_consequents before reduction:", weighted_consequents.shape)

        # Calculate output
        output = np.sum(weighted_consequents, axis=1)

        return output

    def gaussian_mf(self, x, params):
        # Gaussian membership function
        mu, sigma = params
        return np.exp(-0.5 * ((x - mu) / sigma) ** 2)

    def loss(self, X, y):
        # Calculate Mean Squared Error loss
        y_pred = self.forward_pass(X)
        return np.mean((y_pred - y) ** 2)

    def backward_pass(self, X, y):
        # Backpropagation to update parameters
        batch_size = X.shape[0]

        # Compute gradients of consequent parameters
        y_pred = self.forward_pass(X)
        d_consequent_params = np.zeros_like(self.consequent_params)
        for i in range(self.n_rules):
            d_consequent_params[i] = np.sum((y_pred - y) * self.rule_weights[i])

        # Update consequent parameters
        self.consequent_params -= self.learning_rate * d_consequent_params

        # Compute gradients of membership function parameters
        d_mf_params = np.zeros_like(self.mf_params)
        for i in range(self.n_rules):
            for j in range(self.n_inputs):
                d_mf_params[j, i, 0] = np.sum((y_pred - y) * self.rule_weights[i] *
                                              self.consequent_params[i] * (X[:, j] - self.mf_params[j, i, 0]) /
                                              self.mf_params[j, i, 1] ** 2 *
                                              self.gaussian_mf(X[:, j], self.mf_params[j, i]))

                d_mf_params[j, i, 1] = np.sum((y_pred - y) * self.rule_weights[i] *
                                              self.consequent_params[i] *
                                              ((X[:, j] - self.mf_params[j, i, 0]) ** 2 / self.mf_params[j, i, 1] ** 3 - 1) *
                                              self.gaussian_mf(X[:, j], self.mf_params[j, i]))

        # Update membership function parameters
        self.mf_params -= self.learning_rate * d_mf_params

    def fit(self, X, y, epochs=100):
        # Train the ANFIS model
        for epoch in range(epochs):
            self.backward_pass(X, y)
            if epoch % 10 == 0:
                print(f"Epoch {epoch}, Loss: {self.loss(X, y)}")

# Example usage
# Assuming X_train and y_train are your training data

# Reshape X_train if it's a 1D array
X_train = X_train.reshape(-1, 1) if X_train.ndim == 1 else X_train

# Instantiate ANFIS model
anfis_model = ANFIS(n_inputs=X_train.shape[1], n_rules=10, learning_rate=0.01)
# Train the model
anfis_model.fit(X_train, y_train, epochs=100)
```
